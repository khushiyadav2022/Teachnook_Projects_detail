```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/IRIS.csv")
df.head()
```

```python
#input-sepal_length	sepal_width	petal_length	petal_width	
#output-species
```

```python
#2.step no 2 is not required
```

```python
#3.step no 3 is also not required for now
```

```python
#4.Divide the data into i/p and o/p
#input(x) is always 2 dimension
#output(y) is always 1 dimension
x=df.iloc[:,0:4].values
y=df.iloc[:,4].values         #if no ':' then array is 1 dimensional
```

```python
#5. Train and test variables-if you have more data, perform step no 5 for better accuracy
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0) #it make data as constant
#w,x,y,z     variable name you can give like this too
#the variable name can change , but variable order cannot change
```

```python
#After train_test_split
#75% of data in x goes to x_train, rest 25% goes to x_test
#75% of data in y goes to y_train, rest 25% goes to y_test
```

```python
print(x.shape) #100%
print(x_train.shape) #75%
print(y_train.shape) #25%
```

```python
print(y.shape)  #100%
print(y_train.shape) # 75%
print(y_test.shape)  #25%
```

```python
#6. we are not performing step no 6 bcoa all are same decimal so not need to do normalization
```

```python
#7. Run a CLASSIFIER, Regressor or Clusterer
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train) #to train my model, we need i/p training and well as o/p training data
```

```python
#9.PREDICT THE OUTPUT
y_pred=model.predict(x_test) #using the input testing values, we predict the output
y_pred #prediction o/p
```

```python
y_test #actual o/p values
```

```python
#10. Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.amazon.in/Samsung-Phantom-Storage-Additional-Exchange/product-reviews/B09SH994JW/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page=requests.get(url)
page             #try i get 503 error
```

```python
#always ignore top positive and top critical review
#there are 10 reviews in customerreview page
```

```python
page.content
#here we have got the source code of amazon customer review page
#but the source is jumbled and not in readable format
```

```python
soup=bs(page.content,'html.parser')
soup
#html-parser converts the html code into better readable format
```

```python
#Now let us consider the Review Names
names=soup.find_all('span',class_='a-profile-name')[2:]  #top positive and top critical review we delet here by [2:]
names
```

```python
len(names)
```

```python
#Now let us create a dataframe using the data from the names list
import pandas as pd
df=pd.DataFrame(names,columns={'Customer Name'})
df
```

```python
#Next is review title
r_title=soup.find_all('a',class_='review-title')
r_title
```

```python
len(r_title)
```

```python
#Data Cleaning or Data Filtering
review_title=[] #empty list

#for i in range(0,10)
for i in range (0,len(r_title)):
  review_title.append(r_title[i].get_text()[1:-1])       #why using this [1:-1] find out?
review_title
```

```python
#Next is review date
dates=soup.find_all('span',class_='review-date')[2:]
dates
```

```python
len(dates)
```

```python
#Data Cleaning
review_date=[] #empty list
for i in range(0,len(dates)):
  review_date.append(dates[i].get_text().replace('Reviewed in India 🇮🇳 on',''))   #we replace by this 'Reviewed in India 🇮🇳 on' bcoz we want only date 
review_date
```

```python
df['review_title']=review_title
df['review_date']=review_date
df
```

```python
#Assignment
#Review Content
#Review rating
```

```python
r_content=soup.find_all('span',class_='review-text-content')
r_content
```

```python
len(r_content)
```

```python
#Data Cleaning or Data Filtering
review_content=[] #empty list

#for i in range(0,10)
for i in range (0,len(r_content)):
  review_content.append(r_content[i].get_text().replace('\n',''))       #why using this [1:-1] find out?
review_content
```

```python
len(review_content)
```

```python
rating=soup.find_all('i',class_='review-rating')[2:]
```

```python
len(rating)
```

```python
#Data Cleaning
review_rating=[] #empty list
for i in range(0,len(rating)):
  review_rating.append(rating[i].get_text().replace('out of 5 stars',''))   #we replace by this 'Reviewed in India 🇮🇳 on' bcoz we want only date 
review_rating
```

```python
df['review_title']=review_title
df['review_date']=review_date
df['review_content']=review_content
df['review_rating']=review_rating
df
```

```python
#FORECASTING LIBRARY-FBPROPHET
#forcasting-Prediction somthing with respect to time using the past and present data
#forcasting-vegetable prices
#Dataset-nhrdf.org
#1 quintal-100kgs, 1 ton-1000kg
```

```python
#Take the data and create dataframe
import pandas as pd
df=pd.read_html('/content/DailyWiseMarketArrivals.html')[-1]
df  #-1 here consider the table with index -1
```

```python
type(df)
```

```python
df.shape #4189 rows and 6 columns
```

```python
df.size #tells us total no of elements in dataframe
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Now let us drop the null values
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
#FB PROPHET- IT ONLY TAKES DATA IN FORM OF COLUMNS- ds and y -compulsory
#SO rename the date column as ds and rename the modal price(Rs/q) as y
df = df.rename(columns = {'Modal Price (Rs/q)':'y','Date':'ds'})
df
```

```python
#Let us create a new dataframe and consider only 2 column-ds and y from old dataframe
df1 = df[['ds','y']]
df1
```

```python
df1.info()
```

```python
#The datatype of the ds column in object/string format.convert it to datetime format
#the datatype of y column is in object/string format.convert it to interger format
df1['ds']=pd.to_datetime(df1['ds'])
df1['y']=df1['y'].astype(int)
df1.info()
```

```python
#Visulaisation
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5)) #just increasing the figure size
plt.scatter(df1['ds'],df1['y'])
```

```python
#FB PROPHET 
import prophet
op=prophet.Prophet() 
```

```python
op.fit(df1) #we are plotting the value of df1 in Fb prophet library 
```

```python
forecast=op.make_future_dataframe(periods=365,freq='D')
forecast =op.predict(forecast)
```

```python
op.plot(forecast, xlabel='Date',ylabel='Potato prices/q')
#the black dots- Actual ponts of potato prices(1st jan 2005- 17 sep 2022)
#light blue shade -it is the min and max price of potatos as prediction by our model
#Dark blue line is the best fit line
```
